```python
# Código comentado

class Empleados:
    """
    Clase que representa a un empleado de la empresa.

    Atributos:
        nombre (str): Nombre del empleado.
        apellidos (str): Apellidos del empleado.
        edad (int): Edad del empleado.
        posicion (str): Puesto o cargo en la empresa.
        año (int): Año de ingreso a la empresa.
        vacaciones (int): Días de vacaciones disponibles.
        herramientas (list): Lista de herramientas que el empleado sabe usar.
        salario (float): Salario en euros.
    """
    def __init__(self, nombre, apellidos, edad, posicion, año, vacaciones, herramientas, salario):
        """
        Inicializa un objeto Empleados con todos sus atributos.
        """
        self.nombre = nombre
        self.apellidos = apellidos
        self.edad = edad
        self.posicion = posicion
        self.año = año
        self.vacaciones = vacaciones
        self.herramientas = herramientas
        self.salario = salario

    def descripcion(self):
        """
        Devuelve una descripción completa del empleado, incluyendo nombre,
        posición, año de ingreso y días de vacaciones disponibles.

        Returns:
            str: Descripción del empleado.
        """
        # Se construye un string con toda la información relevante
        return (f"El empleado {self.nombre} {self.apellidos} trabaja como {self.posicion} "
                f"desde el año {self.año}. Actualmente le quedan {self.vacaciones} días de vacaciones.")
    
    def cambiando_posicion(self):
        """
        Revisa las herramientas del empleado y da un mensaje personalizado según su conocimiento.

        - Si el empleado sabe Python: felicita.
        - Si solo sabe Excel: sugiere aprender Python.
        - Si no sabe ninguna de estas: motiva a aprender Python.
        """
        # Convertimos todas las herramientas a minúsculas para compararlas sin errores
        herramientas_lower = [x.lower() for x in self.herramientas]

        if "python" in herramientas_lower:
            print(f"¡Enhorabuena {self.nombre} por estar usando Python!")
        elif "excel" in herramientas_lower:
            print(f"¡Deja de usar Excel, {self.nombre}, y ponte a usar Python!")
        else:
            print(f"¡Ponte a usar Python, {self.nombre}!")

    def calcular_antiguedad(self):
        """
        Calcula la antigüedad del empleado desde su año de ingreso hasta la fecha actual
        y la imprime en años, meses y días.
        """
        from datetime import datetime
        from dateutil.relativedelta import relativedelta 
        
        # Fecha actual
        ahora = datetime.now()
        # Fecha de ingreso (primer día del año de ingreso)
        origen = datetime(self.año, 1, 1) 
        # Calculamos la diferencia usando relativedelta para obtener años, meses y días
        diferencia = relativedelta(ahora, origen)
        print(f"Han pasado {diferencia.years} años, {diferencia.months} meses y {diferencia.days} días "
              f"desde que {self.nombre} {self.apellidos} lleva en la empresa")

    def solicitar_aumento_vacaciones(self, dias_solicitados):
        """
        Permite al empleado solicitar un aumento de días de vacaciones.

        Args:
            dias_solicitados (int): Número de días adicionales que desea el empleado.

        Returns:
            int or None: Devuelve la nueva cantidad de días de vacaciones si es válido, None si no.
        """
        if dias_solicitados > 0:
            self.vacaciones += dias_solicitados
            print(f"¡Perfecto! Ahora tienes {self.vacaciones} días de vacaciones.")
            return self.vacaciones
        else:
            # Validación de entrada incorrecta
            print(f"Lo lamento, {dias_solicitados} no es un número válido. Debe ser mayor que 0.")
            return None
    
    def registrar_herramienta(self, herramienta):
        """
        Agrega una herramienta a la lista del empleado si no estaba registrada.
        
        Args:
            herramienta (str): Nombre de la herramienta.
        """
        if herramienta not in self.herramientas:
            self.herramientas.append(herramienta)
            print(f"{herramienta.capitalize()} se ha registrado correctamente.")
        else:
            # Ya existía en la lista
            print(f"Lo siento, {herramienta} ya está en la lista.")

    def cambiar_posicion(self, nueva_posicion, nuevo_salario):
        """
        Permite cambiar la posición y el salario del empleado.

        Args:
            nueva_posicion (str): Nuevo cargo o posición del empleado.
            nuevo_salario (float): Nuevo salario.
        """
        if nuevo_salario <= 0:
            # Validación de salario no válido
            print("El salario debe ser mayor que 0.")
            return
        # Actualizamos atributos
        self.posicion = nueva_posicion
        self.salario = nuevo_salario
        print(f"El salario de {self.nombre} {self.apellidos} es ahora {self.salario}€, "
              f"su nueva posición es {self.posicion.lower()}.")

    def tomar_vacaciones(self):
        """
        Permite al empleado tomar días de vacaciones.

        Valida:
        - Que se ingrese un número válido.
        - Que el número sea positivo.
        - Que no exceda los días de vacaciones disponibles.
        """
        while True:
            try:
                dias = int(input("¿Cuántos días quieres de vacaciones?: "))
            except ValueError:
                print("Por favor, introduce un número válido.")
                continue

            if dias <= 0:
                print("El número tiene que ser mayor que 0.")
            elif dias > self.vacaciones:
                print(f"No puedes tomar más días de los que tienes. Te quedan {self.vacaciones} días.")
            else:
                # Se descuentan los días tomados
                self.vacaciones -= dias
                print(f"Días de vacaciones actualizados. Ahora te quedan {self.vacaciones} días.")
                break    
    
    def solicitar_aumento_salario(self):
        """
        Permite al empleado solicitar un aumento de salario en porcentaje.
        Valida que el porcentaje sea válido y actualiza el salario.
        """
        while True:
            try:
                aumento = float(input("¿En qué porcentaje deseas que se incremente tu salario?"))
            except ValueError:
                print("Por favor, introduce un número válido.")
                continue
                
            if aumento <= 0:
                print("El número tiene que ser mayor que 0")
            else:
                # Calculamos el nuevo salario sumando el porcentaje al actual
                self.salario = ((aumento/100)*self.salario) + self.salario
                print(f"Salario actualizado. Tu salario ahora es {self.salario}€.")
                break
    
    def realizar_evaluacion(self):
        """
        Permite al empleado recibir una calificación de desempeño.
        Valida que la calificación sea una de las opciones correctas y devuelve un mensaje de retroalimentación.

        Returns:
            str: Mensaje con feedback basado en la calificación ingresada.
        """
        opciones = ["excelente", "bueno", "regular", "malo"]
        
        while True:
            # Solicitamos la calificación al usuario
            calificacion = input("Califica tu desempeño (Excelente, Bueno, Regular, Malo): ").strip().lower()
            if calificacion in opciones:
                break
            else:
                # Entrada no válida
                print("Calificación no válida. Por favor ingresa Excelente, Bueno, Regular o Malo.")

        # Generamos mensaje según la calificación
        if calificacion == "excelente":
            mensaje = f"¡Felicidades {self.nombre}! Tu desempeño ha sido excelente. ¡Sigue así!"
        elif calificacion == "bueno":
            mensaje = f"Buen trabajo {self.nombre}, tu desempeño ha sido bueno. Puedes esforzarte aún más."
        elif calificacion == "regular":
            mensaje = f"{self.nombre}, tu desempeño ha sido regular. Hay áreas que puedes mejorar."
        else:  # "malo"
            mensaje = f"{self.nombre}, tu desempeño ha sido malo. Es momento de reflexionar y mejorar."

        return mensaje

```

"If the class you're writing is a specialized version of another class you wrote, you can use *inheritance*. When one class *inherits* from another, it automatically takes on all the attributes and methods of the first class. The original class is called the *parent class*, and the new class is the *child class*. The child class inherits every attribute and method from its parent class but is also free to define new attributes and methods of its own”.

* Implementa los siguientes métodos en la clase "Gerente":
     * `asignar_tareas(tarea, empleado)`: Este método debe recibir el nombre de una tarea y el nombre de un empleado y debe devolver una cadena que indique que el gerente ha asignado la tarea a ese empleado.
     * `revisar_desempeno_empleado(empleado)`: Este método debe recibir el nombre de un empleado y debe devolver una cadena que indique que el gerente está revisando el desempeño de ese empleado.
     * `gestionar_empleados(accion, empleado)`: Este método permite al gerente realizar acciones de contratación o despido de empleados a cargo. Debe recibir una acción ("contratar" o "despedir") y el nombre de un empleado. Si la acción es "contratar", agrega al empleado a la lista de empleados a cargo y devuelve un mensaje indicando que el empleado ha sido contratado por el gerente. Si la acción es "despedir" y el empleado está en la lista de empleados a cargo, lo elimina y devuelve un mensaje de despido. Si la acción no es válida, devuelve un mensaje indicando que la acción no es válida.

```python
class Empleados:
    '''Clase base para empleados'''
    def __init__(self, nombre, apellidos, edad, posicion, año, vacaciones, herramientas, salario):
        self.nombre = nombre
        self.apellidos = apellidos
        self.edad = edad
        self.posicion = posicion
        self.año = año
        self.vacaciones = vacaciones
        self.herramientas = herramientas
        self.salario = salario

    def info(self):
        return (f"{self.nombre} {self.apellidos} | Edad: {self.edad} | Posición: {self.posicion} "
                f"| Año: {self.año} | Vacaciones: {self.vacaciones} días "
                f"| Herramientas: {', '.join(self.herramientas)} | Salario: {self.salario}€")


class Gerente(Empleados):
    '''Clase hija de 'Empleados'. Añade:
      - departamento (str)
      - empleados_a_cargo (list)'''
    def __init__(self, nombre, apellidos, edad, posicion, año, vacaciones, herramientas, salario, departamento, empleados_a_cargo):
        super().__init__(nombre, apellidos, edad, posicion, año, vacaciones, herramientas, salario)
        self.departamento = departamento
        self.empleados_a_cargo = empleados_a_cargo
    
    def info(self):
        empleados = ", ".join(self.empleados_a_cargo) if self.empleados_a_cargo else "Nadie"
        return (f"{super().info()} | Departamento: {self.departamento} "
                f"| Empleados a cargo: {empleados}")

    # Métodos adicionales
    def asignar_tareas(self, tarea, empleado):
        return f"El gerente {self.nombre} ha asignado la tarea '{tarea}' a {empleado}."

    def revisar_desempeno_empleado(self, empleado):
        return f"El gerente {self.nombre} está revisando el desempeño de {empleado}."

    def gestionar_empleados(self, accion, empleado):
        if accion.lower() == "contratar":
            self.empleados_a_cargo.append(empleado)
            return f"{empleado} ha sido contratado por el gerente {self.nombre}."
        
        elif accion.lower() == "despedir":
            if empleado in self.empleados_a_cargo:
                self.empleados_a_cargo.remove(empleado)
                return f"{empleado} ha sido despedido por el gerente {self.nombre}."
            else:
                return f"{empleado} no está en la lista de empleados a cargo."
        
        else:
            return f"Acción '{accion}' no válida. Use 'contratar' o 'despedir'."
```
